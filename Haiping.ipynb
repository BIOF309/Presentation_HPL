```python
# Display how many genes have more than 2 folds change
data_fold = df['Fold_change']
data_qvalue = df['q_value']
fold_cutoff = 1
qvalue_cutoff = 0.05
number_of_genes = len(df.index)

filtered_ids = []
for i in range(0, number_of_genes):
    if (abs(data_fold[i]) >= fold_cutoff) and (data_qvalue[i] <= qvalue_cutoff):
        filtered_ids.append(i)
        
filtered = df.ix[filtered_ids,:]
print("Number of DE genes: ")
print(len(filtered.index))

filtered.head()
```

```python
# Display how many genes have higher expression in STTM396 transgenic tomato Green fruits compared with amiR396
fold_cutoff = 1
qvalue_cutoff = 0.05
number_of_genes = len(df.index)

filtered_ids = []
for i in range(0, number_of_genes):
    if (data_fold[i] >= fold_cutoff) and (data_qvalue[i] <= qvalue_cutoff):
        filtered_ids.append(i)
        
filtered = df.ix[filtered_ids,:]
print("Number of DE genes: ")
print(len(filtered.index))

filtered.head()
```

```python
# Display how many genes have lower expression in STTM396 transgenic tomato Green fruits compared with amiR396
fold_cutoff = 1
qvalue_cutoff = 0.05
number_of_genes = len(df.index)

filtered_ids = []
for i in range(0, number_of_genes):
    if (data_fold[i] < fold_cutoff) and (data_qvalue[i] <= qvalue_cutoff):
        filtered_ids.append(i)
        
filtered = df.ix[filtered_ids,:]
print("Number of DE genes: ")
print(len(filtered.index))

filtered.head()
```

```python
# RNA-seq result follows standard normal distribution (histogram)
data_fold = df['Fold_change']
plt.hist(data_fold,
        bins=100,
        density = True,
        alpha=1,
        color='steelblue',
        edgecolor='darkblue')
plt.xlabel('Fold_change_log2')
plt.ylabel('Gene_density')
plt.show()
```

```python
p1=sns.kdeplot(data_fold, shade=True, label = 'DEG Distribution')
plt.legend(loc='upper left')
plt.xlabel('Fold_change')
plt.ylabel('Gene_density')
plt.show()
```

```python
# Extract gene annotation containing 'SUGAR'
Sugar = df[df['Gene_Annotation'].str.contains("SUGAR")]
Sugar
```

```python
# Extract gene annotation containing 'STARCH'
Starch = df[df['Gene_Annotation'].str.contains("STARCH")]
Starch
```

```python
# Extract gene annotation containing 'SUCROSE'
Sucrose = df[df['Gene_Annotation'].str.contains("SUCROSE")]
Sucrose
```

```python
# Combine the gene list of the above three'Sugar, Starch, Sucrose'
Sugar_gene= pd.concat([Sugar, Starch, Sucrose])
Sugar_gene
```

```python
# write a CSV file from DataFrame
Sugar_gene.to_csv('Sugar_gene.csv')
```

```python
#Get targeted colums (FPKM values for two samples) from "Sugar_gene"
df_FPKM = Sugar_gene[['Gene_ID','amiR396_GF_FPKM','STTM396_GF_FPKM']]
df_FPKM
```

```python
df_FPKM_index = df_FPKM.set_index('Gene_ID')
df_FPKM_index
```

```python
#Heatmap for expression of sweetness-related genes
df_FPKM_index = df_FPKM.set_index('Gene_ID')
sns.clustermap(df_FPKM_index,
              z_score=1,
              metric="euclidean",
              method="ward",
              cmap="Blues",
              annot=True,
              linewidths=1.0)
plt.show()
```
